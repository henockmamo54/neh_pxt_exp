```python
import os
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
path=f"C:/Workplace/C++/d2ome_v2/v2/v2/bin/Debug/"  
protiens=np.unique([x for x in os.listdir(path) if (("_MOUSE.csv" in x) and ("RSS_comparision_" in x))])
print("protiens",len(protiens))

data=pd.DataFrame()
for file in protiens:
    data=pd.concat([data,pd.read_csv(path+file)])
print("peptieds",data.shape)

data=data.dropna()
# data.RSS_FirstThree=data.RSS_FirstThree_unnormalized

# ====================================================
# ====================================================
# ====================================================
print("ALL peptieds",data.shape[0])
print("Abundance < 10^6",100*data[data.Abundance<1E6].shape[0]/data.shape[0])
print("Abundance < 10^7",100*data[data.Abundance<1E7].shape[0]/data.shape[0])
print("Abundance < 10^8",100*data[data.Abundance<1E8].shape[0]/data.shape[0])
print("Abundance < 10^9",100*data[data.Abundance<1E9].shape[0]/data.shape[0])
print("Abundance < 10^10",100*data[data.Abundance<1E10].shape[0]/data.shape[0])

# data=data[data.Abundance<1E7]
```

```python
def closest(lst, K):      
    return lst[min(range(len(lst)), key = lambda i: abs(lst[i]-K))]

data_with_closest_index=data[['mo','io_e','io_a1a0','io_a2a0','io_a2a1','Protein', 'Peptied', 'charge', 'Abundance', 'RSS_compelete','RSS_FirstThree_unnormalized',
       'RSS_FirstThree', 'MAP_C', 'MAP_P', 'L1_distance_C', 'L1_distance_p','sap_c', 'sap_p','RMSE_first_three', 'RMSE_last_three']]
data_with_closest_index['closest_3']=[closest(x[2:5],x[0]) for x in np.array(data_with_closest_index)]
data_with_closest_index['closest']=[closest(x[1:5],x[0]) for x in np.array(data_with_closest_index)]
data_with_closest_index["RMSE_compelete"]=data_with_closest_index.RSS_compelete**0.5
data_with_closest_index['sindex']=[list(x[1:5]).index(x[-2]) for x in np.array(data_with_closest_index)]
print(data.shape)
print(data_with_closest_index.shape)
print("RSS_compelete<RSS_FirstThree",data_with_closest_index[data_with_closest_index.RSS_compelete<data_with_closest_index.RSS_FirstThree].shape[0]/data_with_closest_index.shape[0])
```
